# Interpolation for List

```python
# Xử lí bằng 2 cách linear và nearest

lst_data = [1, 1.1, None, 1.4, None, 1.5, None, None, 2.0]

result = lst_data.copy()
n = len(lst_data)

for i in range(n):
    if result[i] is None:
        left = i - 1
        right = i + 1

        # tìm trái
        while left >= 0 and result[left] is None:
            left -= 1

        # tìm phải
        while right < n and result[right] is None:
            right += 1

        # chọn gtri gần nhất
        if left >= 0 and right < n:
            if i - left <= right - i:
                result[i] = result[left]
            else:
                result[i] = result[right]

print("Inut: ", lst_data)
print("Output: ", result)
```

# FIX NỘI SUY LẤY GIÁ TRỊ GẦN NHẤT



```python
list_data = [None, None, 1, 1.1, None, 1.4, None,  1.5, None, None, None, None, 1.8, 2.0, None, None]

# Copy data
data = list_data.copy()

table = [None] * len(list_data)

# 0 -> None, 1 -> có Gtri, 2 -> từ None sang có gtri
for i, value in enumerate(list_data):
    if value is None:
        table[i] = 0
    else: 
        table[i] = 1

print(table)

for i, value in enumerate(table):
    # Xử lí các giá trị None
    if value == 0:
        left = i - 1
        right = i + 1

        # Tìm giá trị thực bên trái
        while left >= 0 and table[left] != 1:
            left -= 1

        # Tìm giá trị thực bên phải
        while right < len(data) and table[right] != 1:
            right += 1
        
        # Lấy giá trị gần nhất

        #Xử lý khi cả 2 bên đều có giá trị
        if left >= 0 and right < len(data):
            if i - left <= right - i:
                data[i] = data[left]
            else:
                data[i] = data[right]
        # Xử lý khi chỉ có bên trái có giá trị
        elif left >= 0:
            data[i] = data[left]
        # Xử lý khi chỉ có bên phải có giá trị
        elif right < len(data):
            data[i] = data[right]
        
        table[i] = 2

print("Input: " + str(list_data))
print("Output: " + str(data))
```

# NỘI SUY LẤY GIÁ TRỊ TRUNG BÌNH GẦN NHẤT

```python
list_data = [None, None, 1, 1.1, None, 1.4, None,  1.5, None, None, None, None, 1.8, 2.0, None, None]

# Copy data
data = list_data.copy()

table = [None] * len(list_data)

# 0 -> None, 1 -> có Gtri, 2 -> từ None sang có gtri
for i, value in enumerate(list_data):
    if value is None:
        table[i] = 0
    else: 
        table[i] = 1

print(table)

for i, value in enumerate(table):
    # Xử lí các giá trị None
    if value == 0:
        left = i - 1
        right = i + 1

        # Tìm giá trị thực bên trái
        while left >= 0 and table[left] != 1:
            left -= 1

        # Tìm giá trị thực bên phải
        while right < len(data) and table[right] != 1:
            right += 1
        
        # Lấy giá trị gần nhất

        #Xử lý khi cả 2 bên đều có giá trị
        if left >= 0 and right < len(data):
            data[i] = (data[left] + data[right]) / 2
        # Xử lý khi chỉ có bên trái có giá trị
        elif left >= 0:
            data[i] = data[left]
        # Xử lý khi chỉ có bên phải có giá trị
        elif right < len(data):
            data[i] = data[right]
        
        table[i] = 2

print("Input: " + str(list_data))
print("Output: " + str(data))
```
