# Relu

```python
import math

def relu(x):
    return max(x,0)

print("Relu")
print(relu(3))
print(relu(-3))
print(relu(0))
```

# Leaky Relu

```python
def leaky_relu(x, alpha=0.01):
    if x > 0: 
        return x
    else:
        return alpha * x
```

# Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

print("Sigmoid")
print(sigmoid(0))
print(sigmoid(2))
print(sigmoid(-2))
```

# Tanh

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))

print("Tanh")
print(tanh(0))
print(tanh(2))
print(tanh(150))
```

# Elu

```python
def elu(x, alpha=1.0):
    if x > 0:
        return x
    else:
        return alpha * (math.exp(x) - 1)
    
print("Elu")
print(elu(3))
print(elu(-1))
print(elu(0))
```
