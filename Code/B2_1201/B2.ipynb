```python
my_tuple1 = (2, 3)
my_tuple2 = (3, 6)

print(my_tuple1)
print(my_tuple2)
```

# In ra Tổng + Tích 2 vector

```python
def sum_and_prod_manual(t):
    s = 0
    p = 1
    for x in t:
        s += x
        p *= x
    return s, p

s1, p1 = sum_and_prod_manual(my_tuple1)
s2, p2 = sum_and_prod_manual(my_tuple2)

print("Result Sum:", (s1, s2))          # hoặc: tuple([s1, s2])
print("Result Mult:", (p1, p2))         # hoặc: tuple([p1, p2])
```

# Khoảng cách 2 Vector

```python
from math import sqrt,pow

tmp_distance = 0

for i in range(len(my_tuple1)):
    tmp_distance += pow((my_tuple1[i] - my_tuple2[i]), 2)

distance = sqrt(tmp_distance)

print("Distance:", distance)
```
