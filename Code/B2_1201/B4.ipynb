# List Comprehension

```python
corpus = ["I love AI and listen to music"]

stop_word = ["I", "love", "and", "to"]

dictionary = set()

total_list = []

for i in corpus:
    list = i.split()
    total_list += list

for key in stop_word:
    if key in total_list:
        total_list.remove(key)

for i in total_list:
    dictionary.add(i)

dictionary = sorted(dictionary)
    
print(dictionary)
```
