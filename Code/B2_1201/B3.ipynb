# Bag-Of-Word

```python
corpus = ["Tôi thích môn Toán", "Tôi thích AI", "Tôi thích âm nhạc"]

total_list = []
for s in corpus:
    words = s.split()
    total_list += words

dictionary = sorted(set(total_list))

verb = "Tôi thích AI thích toán"
list_verb = verb.split()

binary = dict.fromkeys(dictionary, 0)
for w in list_verb:
    if w in binary:
        binary[w] += 1

vector = [binary[k] for k in dictionary]

print(dictionary)
#print(binary)
print(vector)
```
