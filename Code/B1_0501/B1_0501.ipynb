```python
def giaiThua(n):
    if n <= 1:
        return 1
    else:
        return n * giaiThua(n-1)
    


rs = giaiThua(-1)
print(rs)
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price
    
    def prices(self, capacity):
        return self.price * capacity
    
    def display(self):
        print(f"Product is {self.name} with prices: {self.price} ")

mac = Product("Macbook", 50)

mac.display()
```
